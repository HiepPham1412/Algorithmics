```python
def starry_field(board):
    # this adds 0 to the left, right and top column of the board so easier to code later
    board = ['0'+ i + '0' for i in board]
    board.insert(0, '0'*(len(board) +2))
    # create an nxn list of list each element contains a dictionary storing star level and
    # accumulated value of the path
    board_dim = len(board)
    memo = [[{1:0} for i in range(board_dim+1)] for j in range(board_dim)]
    # loop through each element of each row, then calcualte the possible path
    for i in range(1, board_dim):#
        for j in range(1, board_dim):
            star = 1  if board[i][j] == '*' else 0
            cell_val = 0 if star ==1 else int(board[i][j])
            #left cell of previous row
            inner_tmp = dict()
            left = dict(memo[i-1][j-1])
            for l in list(left.keys()):
                inner_tmp[l] = left[l]+ (l+star)*cell_val
            # middle cell
            middle = dict(memo[i-1][j])
            for m in list(middle.keys()):
                if m in inner_tmp.keys():
                    if middle[m] + (m+star)*cell_val > inner_tmp[m]:
                        inner_tmp[m] = middle[m] + (m+star)*cell_val
                else:
                    inner_tmp[m] = middle[m]+ (m+star)*cell_val
            #right conner cell
            right = dict(memo[i-1][j+1])
            for r in list(right.keys()):
                if r in inner_tmp.keys():
                    if right[r] + (r+star)*cell_val > inner_tmp[r]:
                        inner_tmp[r] = right[r] + (r+star)*cell_val
                else:
                    inner_tmp[r] = right[r]+ (r+star)*cell_val
            # remove all NON_POTENTIAL path where accumulated value of lower star level
            # is lower than that of the higher star level
            final = dict()
            keys = sorted(inner_tmp.keys(), reverse = True)
            final[keys[0]+star] = inner_tmp[keys[0]]
            for f in range(1,len(inner_tmp)):
                tmpt = inner_tmp[keys[0]]
                if inner_tmp[keys[f]] > tmpt:
                    final[keys[f] + star] = inner_tmp[keys[f]]
                    tmpt = inner_tmp[keys[f]]
            memo[i][j] = dict(final)
    # get the MAX value from the list of dictionary in the last row
    final_row_dict = memo[board_dim-1]
    print(final_row_dict)
    final_row_list = []
    for i in final_row_dict:
        final_row_list.append([int(j) for j in i.values()])
    Max = 0
    for i in final_row_list:
        Max = max(Max, max(i))
  # implement your solution here
    return Max

board = [
  "0587828028", 
  "2967480535", 
  "82*14606*4", 
  "*986777360", 
  "0*85524798", 
  "059*736685", 
  "084821304*", 
  "15880495*6", 
  "291*994640", 
  "13*412531*", 
  ]
  
expected_result = 149

my_result = starry_field(board)
print("Your result is: ", my_result)

if my_result == expected_result:
  print("Result correct!")
else:
  print("Result incorrect!")
```
